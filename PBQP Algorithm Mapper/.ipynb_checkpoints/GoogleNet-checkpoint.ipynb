```python
import math
import numpy as np

bandwidth=64 * 10**6  #GB/ms
freq=300 * 10**3 #cycles/ms


#fix these load latencies! wrong for im2col right now
def im2col_OS_latency(a,b,c,P1,P2): #a: H^2, b:K^2*C_in, c:C_out
	return math.ceil(a/P1)*math.ceil(c/P2)*b/freq + (a*b+b*c)/bandwidth

def im2col_WS_latency(a,b,c,P1,P2):
	return math.ceil(b/P1)*math.ceil(c/P2)*a/freq + (a*b+b*c)/bandwidth

def im2col_IS_latency(a,b,c,P1,P2):
	return math.ceil(b/P1)*math.ceil(a/P2)*c/freq + (a*b+b*c)/bandwidth
# def kn2row_OS_latency(a,b,c,k1,k2,P1,P2): #a: H^2, b:C_in, c:C_out
# 	return math.ceil(b/P1)*math.ceil(c/P2)*a*k1*k2/freq + (a*b+b*c)/bandwidth

# def kn2row_WS_latency(a,b,c,k1,k2,P1,P2):
# 	return math.ceil(b/P1)*math.ceil(c/P2)*a*k1*k2/freq + (a*b+b*c)/bandwidth

def kn2row_OS_latency(a,b,c,k,P1,P2): #a: H^2, b:C_in, c:C_out
	return math.ceil(a/P1)*math.ceil(c/P2)*b*k*k/freq

def kn2row_WS_latency(a,b,c,k,P1,P2):
	return math.ceil(b/P1)*math.ceil(c/P2)*a*k*k/freq

def kn2row_IS_latency(a,b,c,k,P1,P2):
	return math.ceil(b/P1)*math.ceil(a/P2)*c*k*k/freq
#conv_blocks.append((wino_OS_latency(a,b,c,m,r,P1,P2)*(3/r)**2+(3/r)**2*(m+r-1)**2*max(m,r))/freq + (a*b/m**2+b*c)*(m+r-1)**2/bandwidth)
# def wino_OS_latency(a,b,c,k,m,r,P1,P2): #a: H^2, b:C_in, c:C_out
#     #matrix mult time + linear transformation time + load time
# 	return (math.ceil((a/m**2)/P1)*math.ceil(c/P2)*a*(m+r-1)**2 *(k/r)**2 + (k/r)**2*(m+r-1)**2*max(m,r))/freq + (a*b/m**2+b*c)*(m+r-1)**2/bandwidth

# def wino_WS_latency(a,b,c,k,m,r,P1,P2):
# 	return (math.ceil(b/P1)*math.ceil(c/P2)*(a/m**2)*(m+r-1)**2 *(k/r)**2 + (k/r)**2*(m+r-1)**2*max(m,r))/freq + (a*b/m**2+b*c)*(m+r-1)**2/bandwidth

def wino_OS_latency(a,b,c,k,m,r,P1,P2): #a: H^2, b:C_in, c:C_out
    #matrix mult time + linear transformation time + load time
	return (math.ceil((a/m**2)/P1)*math.ceil(c/P2)*b*(m+r-1)**2 *(k/r)**2 + (k/r)**2*(m+r-1)**2*max(m,r))/freq

def wino_WS_latency(a,b,c,k,m,r,P1,P2):
	return (math.ceil(b/P1)*math.ceil(c/P2)*(a/m**2)*(m+r-1)**2 *(k/r)**2 + (k/r)**2*(m+r-1)**2*max(m,r))/freq

def wino_IS_latency(a,b,c,k,m,r,P1,P2):
	return (math.ceil(b/P1)*math.ceil((a/m**2)/P2)*c*(m+r-1)**2 *(k/r)**2 + (k/r)**2*(m+r-1)**2*max(m,r))/freq


def inception_latency(C_in,o,C_out,num_1by1,num_3by3,num_5by5,pool_proj,reduce_3by3,reduce_5by5,P1,P2,df):
    assert(C_out==num_1by1+num_3by3+num_5by5+pool_proj)
    conv_blocks=[[],[],[],[],[],[]]
    num_ops=0
    # From left to right
    # 1 by 1 CONV block
    a=o*o
    b=C_in
    c=num_1by1
    num_ops+=a*b*c
    conv_blocks[0].append(min(im2col_OS_latency(a,b,c,P1,P2),im2col_WS_latency(a,b,c,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))
    conv_blocks[0].append(min(kn2row_OS_latency(a,b,c,1,P1,P2),kn2row_WS_latency(a,b,c,1,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))

    # 1 by 1 CONV block before 3x3
    a=o*o
    b=C_in
    c=reduce_3by3
    num_ops+=a*b*c
    conv_blocks[1].append(min(im2col_OS_latency(a,b,c,P1,P2),im2col_WS_latency(a,b,c,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))
    conv_blocks[1].append(min(kn2row_OS_latency(a,b,c,1,P1,P2),kn2row_WS_latency(a,b,c,1,P1,P2),kn2row_IS_latency(a,b,c,1,P1,P2)))
    
    # 3 by 3 CONV block
    a=o*o
    b=3*3*reduce_3by3
    c=num_3by3
    num_ops+=a*b*c
    conv_blocks[2].append(min(im2col_OS_latency(a,b,c,P1,P2),im2col_WS_latency(a,b,c,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))
    conv_blocks[2].append(min(kn2row_OS_latency(a,b,c,3,P1,P2),kn2row_WS_latency(a,b,c,3,P1,P2),kn2row_IS_latency(a,b,c,3,P1,P2)))
    conv_blocks[2].append(min(wino_OS_latency(a,b,c,3,2,3,P1,P2),wino_WS_latency(a,b,c,3,2,3,P1,P2),wino_IS_latency(a,b,c,3,2,3,P1,P2)))
    
    # 1 by 1 CONV block before 5x5
    a=o*o
    b=C_in
    c=reduce_5by5
    num_ops+=a*b*c
    conv_blocks[3].append(min(im2col_OS_latency(a,b,c,P1,P2),im2col_WS_latency(a,b,c,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))
    conv_blocks[3].append(min(kn2row_OS_latency(a,b,c,1,P1,P2),kn2row_WS_latency(a,b,c,1,P1,P2),kn2row_IS_latency(a,b,c,1,P1,P2)))
    
    # 5 by 5 CONV block
    a=o*o
    b=5*5*reduce_5by5
    c=num_5by5
    num_ops+=a*b*c
    conv_blocks[4].append(min(im2col_OS_latency(a,b,c,P1,P2),im2col_WS_latency(a,b,c,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))
    conv_blocks[4].append(min(kn2row_OS_latency(a,b,c,5,P1,P2),kn2row_WS_latency(a,b,c,5,P1,P2),kn2row_IS_latency(a,b,c,5,P1,P2)))
    conv_blocks[4].append(min(wino_OS_latency(a,b,c,5,2,5,P1,P2),wino_WS_latency(a,b,c,5,2,3,P1,P2),wino_IS_latency(a,b,c,5,2,3,P1,P2)))
    
    # 1 by 1 CONV block after Pool
    a=o*o
    b=C_in
    c=pool_proj
    num_ops+=a*b*c
    conv_blocks[5].append(min(im2col_OS_latency(a,b,c,P1,P2),im2col_WS_latency(a,b,c,P1,P2),im2col_IS_latency(a,b,c,P1,P2)))
    conv_blocks[5].append(min(kn2row_OS_latency(a,b,c,1,P1,P2),kn2row_WS_latency(a,b,c,1,P1,P2),kn2row_IS_latency(a,b,c,1,P1,P2)))
    # print("num_ops:",num_ops/1e6)
    return conv_blocks
```
